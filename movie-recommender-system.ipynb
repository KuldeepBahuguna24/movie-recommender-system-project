```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
 movies=movies.merge(credits,on='title')   #merge movies and credits on the basis of title and store in movies
```

```python
movies.info()
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
#these methods are usefull so we will go throw these to reccomend

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
 movies.head()
```

```python
movies.isnull().sum()  #this line will check the missing data (if present)
```

```python
movies.dropna(inplace=True)   #remove the missing data from series
```

```python
movies.duplicated().sum()
```

```python
movies = movies.drop_duplicates()   #to remove duplicate data
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    #in this function literal_eval will convert the list of string into list
```

```python
movies['genres']=movies['genres'].apply(convert)   # will convert the genres into a simple list 
```

```python
movies['keywords']=movies['keywords'].apply(convert)  #   will convert the keywords into a simple list 
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
# this function will change cast into form of list and will show only top 3 actor names
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# This function will find the director and then stop
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# converting overview from sting to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
# lambda split function will split the content into list
```

```python
movies.head()
```

```python
# we have to remove spaces in between name of genres/keywords/cast and crews so then we can reccomend a proper name

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
# concatinating all infos...
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# creating a new data fram
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
# converting the list of tags into string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#  ['Love','Loving','Loved']
# ['Love','Love','Love']
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# convert all tags into lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# it will store only sort from of words like loved=love..
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
# text vectrization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# use predefine module sklearn for finding the similarity angle between tags
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
cosine_similarity(vectors).shape
```

```python
# if someone will send a movie then this function will provide them 5 similar movie

def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]   # this line will find the movie index
    
    distances = similarity[movie_index]   # distance between all movies
    
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    # enumerate function will change list into tupple  and sort it then reverse it and return 1 to 6 similar movies..

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
